```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('breast-cancer-wisconsin.csv')
```

```python
df
```

```python
df_new = df.rename(columns={'1000025': 'Sample', '5': 'Clump Thickness' , '1': 'Cell size', '1.1': 'cell shape' , '1.2': 'Adhesion' , '2': 'Epithelial Cell' , '1.3': 'Bare nuclei' , '3': 'Bland Chromatin' , '1.4': 'Normal nucleoli' , '1.5': 'Mitoses' , '2.1': 'Class'  })
```

```python
df_new
```

```python
df_new.info()
```

```python
delete = df_new[df_new['Bare nuclei']=='?'].index
df_new.drop(delete,axis=0,inplace=True)
```

```python
df_new.isnull().sum()
```

```python
label_2 = df_new[df_new['Class']==2]
label_4 = df_new[df_new['Class']==4]
```

```python
for i in range(0,10):
  for j in range(i+1, 10):
    sns.scatterplot(x=df_new.columns[i],y=df_new.columns[j],data=label_2,color='blue',alpha=0.3);
    sns.scatterplot(x=df_new.columns[i],y=df_new.columns[j],data=label_4,color='red',alpha=0.3);
    plt.show()
```
